```python
#parte buona
import cv2
import mediapipe as mp
import pandas as pd
import numpy as np
import math
import os
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter, find_peaks

# Percorso video
video_path = "/kaggle/input/squat-validi/VID-20251028-WA0002.mp4"
assert os.path.exists(video_path), "‚ùå File video non trovato!"
print("üé• Video trovato correttamente!")
```

```python
import pandas as pd
import numpy as np

# === Carica il CSV originale con tutti i landmarks estratti ===
df = pd.read_csv("/kaggle/working/landmarks_squat.csv")

# === Landmark chiave: anche e ginocchia ===
key_ids = [23, 24, 25, 26]  # 23=anca sx, 24=anca dx, 25=ginocchio sx, 26=ginocchio dx

# Filtra solo i landmarks chiave
df_key = df[df["landmark_id"].isin(key_ids)].copy()

# Inizializza struttura per tracking ‚Äúsoft‚Äù frame per frame
tracked_points = []
prev_positions = {}

# Soft tracking: mantiene i punti stabili, evita salti enormi
max_jump = 0.15  # soglia massima di movimento tra frame consecutivi (in coordinate normalizzate)

for frame_id in sorted(df_key["frame"].unique()):
    frame_data = df_key[df_key["frame"] == frame_id].copy()
    
    for _, row in frame_data.iterrows():
        lm_id = row["landmark_id"]
        pos = np.array([row["x"], row["y"], row["z"]], dtype=np.float32)
        vis = row["visibility"]
        
        if lm_id in prev_positions:
            dist = np.linalg.norm(pos - prev_positions[lm_id])
            if dist > max_jump or vis < 0.5:
                # Mantieni la posizione precedente se il salto √® troppo grande o visibilit√† bassa
                pos = prev_positions[lm_id]
        
        prev_positions[lm_id] = pos
        tracked_points.append({
            "frame": frame_id,
            "landmark_id": lm_id,
            "x": float(pos[0]),
            "y": float(pos[1]),
            "z": float(pos[2]),
            "visibility": float(vis)
        })

# Salva il CSV con solo landmarks chiave e tracking stabilizzato
df_tracked = pd.DataFrame(tracked_points)
df_tracked.to_csv("/kaggle/working/landmarks_key_tracked.csv", index=False)
print(f"‚úÖ Landmark chiave tracciati: {len(df_tracked)} punti in {df_tracked['frame'].nunique()} frame")
```

```python
# Seleziona le anche (ID 23 e 24)
hip_left = df[df["landmark_id"] == 23]
hip_right = df[df["landmark_id"] == 24]

# Media verticale del bacino
df_hips = pd.DataFrame({
    "frame": hip_left["frame"].values,
    "y_mean": (hip_left["y"].values + hip_right["y"].values) / 2
})

# Filtro di smoothing per eliminare rumore
df_hips["y_smooth"] = savgol_filter(df_hips["y_mean"], window_length=15, polyorder=3)
```

```python
from scipy.signal import savgol_filter, find_peaks
import matplotlib.pyplot as plt

# Amplifica il segnale per renderlo pi√π leggibile
df_hips["y_scaled"] = df_hips["y_mean"] * 1000

# Smussa (smoothing)
df_hips["y_smooth"] = savgol_filter(df_hips["y_scaled"], window_length=25, polyorder=3)

# Inverte per cercare i fondi squat
df_hips["y_inv"] = -df_hips["y_smooth"]

# Trova i minimi con parametri pi√π permissivi
minimi, props = find_peaks(
    df_hips["y_inv"],
    distance=50,        # distanza minima in frame tra due squat
    prominence=0.2,     # ora leggendo il segnale amplificato
    width=10
)

print(f"üèãÔ∏è‚Äç‚ôÇÔ∏è Squat rilevati automaticamente: {len(minimi)}")

# Grafico di verifica
plt.figure(figsize=(12, 5))
plt.plot(df_hips["frame"], df_hips["y_smooth"], color='orange', label="Bacino (smussato)")
plt.scatter(df_hips["frame"].iloc[minimi], df_hips["y_smooth"].iloc[minimi],
            color='red', s=80, label="Fondo squat")
plt.legend()
plt.xlabel("Frame")
plt.ylabel("Altezza bacino (scala x1000)")
plt.title("Rilevamento automatico delle ripetizioni di squat (segnale amplificato)")
plt.show()
```

```python
import os
import cv2
import numpy as np
import pandas as pd
import mediapipe as mp
from collections import deque
from scipy.signal import savgol_filter

# === PARAMETRI MODIFICABILI ===
video_path = "/kaggle/input/squat-validi/VID-20251028-WA0002.mp4"
output_path = "/kaggle/working/landmarks_stabilized_v4.mp4"
csv_out = "/kaggle/working/landmarks_stabilized_v4.csv"

roi_fraction_initial = 0.6
visibility_threshold = 0.55
ema_alpha = 0.2           # peso EMA pi√π basso per maggiore smoothing
max_move_threshold = 0.12
roi_safety_margin = 0.15
roi_adapt_start_frame = 50
key_ids = [23, 24, 25, 26]      # anche e ginocchia
connections_lower = [(23,25), (24,26), (23,24)]
smooth_window = 5                 # media velocit√†

assert os.path.exists(video_path), "‚ùå File video non trovato!"
print("üé• Video caricato correttamente")

mp_pose = mp.solutions.pose
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS) or 25.0
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

roi_w = int(width * roi_fraction_initial)
mask = np.zeros((height, width), dtype=np.uint8)
cv2.rectangle(mask, (0,0), (roi_w, height), 255, -1)

prev_landmarks = None
stabilized_records = []
frame_idx = 0
roi_dynamic = [0, roi_w]
vel_history = {i: deque(maxlen=smooth_window) for i in key_ids}

with mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.55, min_tracking_confidence=0.65) as pose:
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_idx += 1

        frame_masked = cv2.bitwise_and(frame, frame, mask=mask)
        rgb = cv2.cvtColor(frame_masked, cv2.COLOR_BGR2RGB)
        results = pose.process(rgb)

        if not results.pose_landmarks:
            lms_arr = prev_landmarks.copy() if prev_landmarks is not None else None
            if lms_arr is None:
                out.write(frame)
                continue
        else:
            lms = results.pose_landmarks.landmark
            lms_arr = np.array([[lm.x, lm.y, lm.z, lm.visibility] for lm in lms], dtype=np.float32)

        # Controllo visibilit√† dei keypoints
        key_vis = lms_arr[key_ids, 3]
        if np.mean(key_vis) < visibility_threshold and prev_landmarks is not None:
            lms_arr = prev_landmarks.copy()

        # Stabilizzazione landmarks
        if prev_landmarks is None:
            prev_landmarks = lms_arr.copy()
        else:
            new_lm = lms_arr.copy()
            for i in key_ids:
                dx = new_lm[i,0] - prev_landmarks[i,0]
                dy = new_lm[i,1] - prev_landmarks[i,1]
                dist = np.hypot(dx, dy)

                if dist > max_move_threshold and new_lm[i,3] < 0.7:
                    new_lm[i] = prev_landmarks[i]
                else:
                    vel = new_lm[i, :3] - prev_landmarks[i, :3]
                    vel_history[i].append(vel)
                    vel_avg = np.mean(vel_history[i], axis=0)
                    pred = new_lm[i, :3] + 0.3 * vel_avg
                    # EMA per stabilizzazione
                    new_lm[i, :3] = (1-ema_alpha)*prev_landmarks[i, :3] + ema_alpha*pred
                    new_lm[i, 3] = max(prev_landmarks[i,3], new_lm[i,3])
            prev_landmarks = new_lm

        # Aggiornamento ROI dinamico
        if frame_idx > roi_adapt_start_frame:
            xs = prev_landmarks[key_ids, 0]
            xmin, xmax = np.min(xs), np.max(xs)
            xmin = max(0, xmin - roi_safety_margin)
            xmax = min(1, xmax + roi_safety_margin)
            roi_dynamic = [int(xmin*width), int(xmax*width)]
            mask[:] = 0
            cv2.rectangle(mask, (roi_dynamic[0], 0), (roi_dynamic[1], height), 255, -1)

        # Disegna connessioni e punti
        for (a,b) in connections_lower:
            pa, pb = prev_landmarks[a], prev_landmarks[b]
            if pa[3] > 0.3 and pb[3] > 0.3:
                x1, y1 = int(pa[0]*width), int(pa[1]*height)
                x2, y2 = int(pb[0]*width), int(pb[1]*height)
                cv2.line(frame, (x1,y1), (x2,y2), (0,0,255), 2)
        for i in key_ids:
            x, y, z, vis = prev_landmarks[i]
            if vis > 0.3:
                cv2.circle(frame, (int(x*width), int(y*height)), 5, (0,255,0), -1)

        cv2.rectangle(frame, (roi_dynamic[0],0), (roi_dynamic[1],height), (255,255,255), 1)
        cv2.putText(frame, f"Frame {frame_idx}", (30,40), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,0), 2)

        # Salvataggio temporaneo
        for idx in key_ids:
            x, y, z, vis = prev_landmarks[idx]
            stabilized_records.append({
                "frame": frame_idx, "landmark_id": idx,
                "x": float(x), "y": float(y), "z": float(z), "visibility": float(vis)
            })

        out.write(frame)
        if frame_idx % 200 == 0:
            print(f"üß© Processati {frame_idx} frame...")

cap.release()
out.release()

# === Applicazione filtro Savitzky-Golay finale su y dei keypoints ===
df_stable = pd.DataFrame(stabilized_records)
for idx in key_ids:
    mask = df_stable["landmark_id"] == idx
    for col in ["x","y","z"]:
        df_stable.loc[mask, col] = savgol_filter(df_stable.loc[mask, col], window_length=7, polyorder=3)

df_stable.to_csv(csv_out, index=False)
print("‚úÖ Video salvato in:", output_path)
print("‚úÖ CSV salvato in:", csv_out)
```

```python
import os
import cv2
import numpy as np

# === Parametri di base ===
video_path = "/kaggle/input/squat-validi/VID-20251028-WA0002.mp4"
output_dir = "/kaggle/working/squat_frames_statici"
os.makedirs(output_dir, exist_ok=True)

# === Funzione per disegno linee e etichette ===
def draw_squat_analysis(frame, y_anca, y_ginocchio, valido):
    h, w = frame.shape[:2]

    if np.isnan(y_anca) or np.isnan(y_ginocchio):
        return frame  # evita errori se dati mancanti

    y_anca_px = int(y_anca * h)
    y_gin_px = int(y_ginocchio * h)

    color = (0, 255, 0) if valido else (0, 0, 255)
    label = "VALIDO ‚úÖ" if valido else "ALTO ‚ùå"

    # linee orizzontali
    cv2.line(frame, (0, y_anca_px), (w, y_anca_px), color, 2)
    cv2.line(frame, (0, y_gin_px), (w, y_gin_px), (255, 255, 0), 2)

    # etichetta testo
    cv2.putText(frame, label, (30, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 3, cv2.LINE_AA)

    # differenza visiva
    cv2.putText(frame, f"Anca: {y_anca:.3f} | Ginocchio: {y_ginocchio:.3f}",
                (30, h - 40), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)
    return frame


# === Carica video ===
cap = cv2.VideoCapture(video_path)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"üé• Frame totali: {total_frames}")

# === Crea mappa frame ‚Üí dati squat ===
# (usa direttamente i frame_fondo gi√† calcolati in df_squat_eval)
frame_targets = {int(row.frame_fondo): row for _, row in df_squat_eval.iterrows()}
print(f"üì∏ Squat da visualizzare: {len(frame_targets)}")

# === Estrai e salva immagini ===
frame_idx = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_idx += 1

    if frame_idx in frame_targets:
        row = frame_targets[frame_idx]
        frame_annotato = draw_squat_analysis(
            frame.copy(),
            row["y_anca_mean"],
            row["y_ginocchio_mean"],
            row["valido"]
        )

        filename = f"squat_{frame_idx}_{'valido' if row['valido'] else 'alto'}.jpg"
        path = os.path.join(output_dir, filename)
        cv2.imwrite(path, frame_annotato)
        print(f"üíæ Salvata immagine: {path}")

cap.release()
print("‚úÖ Tutte le immagini salvate in:", output_dir)
```
